### Data Distributions
    1. uniform Distributions(data points will be equally distributed in any point of data set)
    (density of data will remain equal in every portion of the data set)
    2. Normal Distributions(Gaussian Distribution)

#### To represent distributions we use histogram

```python
data = np.random.randint(1, 100, 1000)
# this function generates data using unform distribution
```

```python
plt.figure(figsize=(8, 3))
plt.hist(data, bins=30)
plt.show()
```

```python
height = np.random.normal(size = 1000)
# normal fucntion generates standard normal distribution i.e.. mean = 0 and std = 1
plt.figure(figsize=(8, 3))
plt.hist(height, bins = 30,ec='k', label="standard normal distribution")
plt.legend()
plt.show()
```

```python
height = np.random.normal(160, 20, 1000)
# normal distribution where mean can take any value 
# normal takes three parameters normal(mean value, deviation from mean value, tatal no of values)
plt.figure(figsize=(8, 3))
plt.hist(height, bins=30, ec='k', label="normal distribution")
plt.legend()
plt.show()
```

```python
# left skew(mean < meadian)
new_data = np.append(height, np.random.uniform(50, 100, 50))
new_data = np.append(new_data, np.random.uniform(100, 130, 50))
plt.hist(new_data, bins=30, label="left skew normal distribution", ec='k')
bar = new_data.mean()
mu = np.median(new_data)
plt.vlines(bar, 0, 140, label="mean", color="red")
plt.vlines(mu, 0, 140, label="median", color="yellow")
plt.legend()
plt.show()
```
