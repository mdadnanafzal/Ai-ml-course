```python
for team in ipl["Team"].unique():
    if team=="Unsold":
        continue
    team_df = ipl[ipl["Team"]==team]
    n = team_df.shape[0]
    p = round(team_df["Price Cr"].sum(), 2)
    print(f"Team {team:<30} has spend {p:>7} Cr rupees to by {n:^5} players.")
```
