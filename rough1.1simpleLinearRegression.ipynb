```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
mpg = sns.load_dataset("mpg").dropna()
```

```python
mpg.head()
```

```python
num_features = ["displacement", "horsepower", "weight", "acceleration"]  # load multiple cols at a time 
x = mpg[num_features]
y = mpg["mpg"]
```

```python
x.head()
```

```python
y.head()
```

```python
x.corrwith(y)  # x.corrwith(y) function gives corelation of y with every x  
```

```python
sns.scatterplot(data = mpg, x = "horsepower", y = "horsepower")
```

```python
np.corrcoef(mpg["horsepower"], mpg["horsepower"])
```

```python
np.corrcoef(mpg["horsepower"], mpg["horsepower"])[0,1]
```

```python
r = np.corrcoef(mpg["horsepower"], mpg["horsepower"])[0,1]
sns.scatterplot(data = mpg, x = "horsepower", y = "horsepower", label =f"r={r:.2f}")
plt.legend()
```

```python
r = np.corrcoef(mpg["horsepower"], mpg["mpg"])[0,1]
mpg.plot(kind="scatter",
        x = "horsepower", 
        y = "mpg",
        label = f"r={r:.2f}")
plt.show()
```

```python
fig, ((ax1, ax2), (ax3, ax3)) = plt.subplots(nrows=2, ncols=2)
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))
seq = zip(["displacement", "horsepower", "weight", "acceleration"],
          [   ax1,             ax2,         ax3,         ax4])
for col, ax in seq:
     r = np.corrcoef(mpg[col], mpg["mpg"])[0,1]
     ax.scatter(mpg[col], mpg["mpg"], label=f"r = r{r:.2f}")   
     ax.set_title(f"{col} vs mpg")
     ax.legend()
 

plt.show()
     
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))
seq = zip(["displacement", "horsepower", "weight", "acceleration"],
          [   ax1,             ax2,         ax3,         ax4])
for col, ax in seq:
     r = np.corrcoef(mpg[col], mpg["mpg"])[0,1]
     ax.plot(mpg[col], mpg["mpg"], "bo", label = f"r = r{r:.2f}", mfc = 'white', alpha=0.5)  
     ax.set_title(f"{col} vs mpg")
     ax.legend()
 

plt.show()
```
