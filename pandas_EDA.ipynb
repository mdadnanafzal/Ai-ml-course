## imp
    autopct shows the percentage of each division of the pie chart ("%.2f%%")
    elplode is used to move the pie chart so that it can be highlighted
    explode=[0.3, 0, 0.3, 0.3, 0] 0.3 means  explode the part 3 by 3%
    

```python
mpg["cylinders"].value_counts().sort_index().plot(kind="pie", autopct="%.2f%%", explode=[0.3, 0, 0.3, 0.08, 0], 
                                                  colors=["gray", "salmon","gray", "gray", "salmon"], shadow=True, 
                                                 counterclock=False, startangle=90) # using counter clock function it reverses the way in which data is shown part by part
```

## pie chart using matplotlib

```python
data = mpg["cylinders"].value_counts()
data
```

```python
data = mpg["cylinders"].value_counts().sort_index().reset_index()
data.columns=["cylinders", "frequency"]
data
```

```python
plt.pie(data["frequency"], labels=data["cylinders"], autopct="%.2f%%", explode=[0, 0.05, 0, 0, 0.05], 
                                                  colors=["gray", "salmon","gray", "gray", "salmon"], shadow=True, 
                                                 labeldistance=0.9,pctdistance=1.2, radius=1.2, startangle=90, 
                                                counterclock=False)
ax = plt.gca()
circle = patches.Circle((0,0), 0.8, facecolor='white')
ax.add_artist(circle)
ax.set_title("distribution of cylinders", pad=60)
plt.show()
```

```python
fd = {"fontsize":15, "fontfamily":"monospace"}  # we can define fd once and use in the fontdict multiple times
```

```python
plt.figure(figsize=(10, 4))
sns.countplot(data=tips, x = "sex", palette="coolwarm",hue="smoker", alpha=1)

ax = plt.gca()  # gca() stands for get current axis
ax.set_title("distribution of data", color="salmon", fontdict={"fontsize":20, "fontweight":"bold", "fontfamily":"monospace"})

ax.set_yticks(np.arange(0, 110, 10))
ax.set_xticklabels(["Male", "Female"], color="red")

ax.set_xlabel("sex", fontdict=fd, color="black")
ax.set_ylabel("frequency", fontdict=fd, color="black")

ax.grid(alpha=1, ls="--")  # ls can be used to set the type of line we want in grid lines

ax.spines["top"].set_visible(False)# this function removes the top spines(lines) of figure
ax.spines["right"].set_visible(False)# this function removes the right spines(lines) of figure
ax.spines["bottom"].set_color("red")
ax.spines["bottom"].set_lw(2)  # lw stands for line width
```

### but there is so many data so it is very difficult to read the graph so we use .sample(frac="percetage")
### to get only some fraction of the actual data set

```python
sample = diamond.sample(frac=0.25)
```

```python
sns.scatterplot(data=sample, x="carat", y = "price", hue="color", palette="rainbow", alpha=0.1)
plt.yscale("log")
plt.show()
```

```python
sns.scatterplot(data=sample, x="carat", y = "price", hue="clarity", palette="rainbow", alpha=1)
plt.yscale("log")
plt.show()
```

```python
sns.scatterplot(data=sample, x="carat", y = "price", hue="clarity", size="cut", palette="rainbow", alpha=0.5)
plt.yscale("log")
plt.show()
```

```python
sns.scatterplot(data=sample, x="carat", y = "price", hue="clarity", size="cut", palette="rainbow", style="color", alpha=0.5)
plt.yscale("log")
plt.show()
```
