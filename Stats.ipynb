```python
import numpy as np
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [4, 6, 5, 3, 6, 7, 5, 3]

m = np.mean(y)
md=np.median(y)
plt.plot([min(x), max(x)], [md, md], color="orange", label="median")
plt.plot([min(x), max(x)], [m, m], color = "red", label = "mean")
for x1, y1 in zip(x, y):
    plt.plot([x1, x1], [y1, m], '--c')
plt.plot([], [], '--c', label="deviation")    
plt.scatter(x, y, label="Scatter")# instead of lines it prints the points only
plt.legend()
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("demonstration of mean line over scatter data")
plt.show()
```

```python
def absolute_deviation(y):
    mu = np.mean(y)
    y = np.array(y)  # so that operation takes element wise 
    mu - y
    return sum(abs(mu-y))/len(y)


y = [10, 40, 30, 50, 20, 60, 70, 80, 30, 80, 70, 60, 50, 60, 6, 10, 12, 16, 20, 30, 22, 25, 10, 43, 44, 45, 35, 38
     , 40, 45, 42, 45,43, 44, 45, 35, 38, 40, 45, 42, 45,43, 44, 45, 35, 38, 40, 45, 42, 45,]
x = np.arange(len(y))
plt.scatter(x, y)
mu  = np.mean(y)
plt.plot([min(x), max(x)], [mu,mu], label="mean", color="red")
abs_dev = absolute_deviation(y)
plt.plot([min(x), max(x)], [mu+abs_dev, mu+abs_dev], label="abs dev", color="yellow")
plt.plot([min(x), max(x)], [mu-abs_dev, mu-abs_dev],   color="yellow")
std = np.std(y)
plt.plot([min(x), max(x)], [mu+std, mu+std], label="std dev", color="magenta")
plt.plot([min(x), max(x)], [mu-std, mu-std], color="magenta")
plt.legend()
```
