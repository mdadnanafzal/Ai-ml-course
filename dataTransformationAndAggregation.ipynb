### Analysis of tip given by different gender

```python
tips["tip"].describe()
```

```python
tips["sex"].describe()
```

DataFrame(male, female) ---> select ---> DataFrame(male)

```python
male_df=tips[tips["sex"] =="Male"]  # all the rows where gender is male is selected
```

```python
print(male_df)
```

```python
female_df = tips[tips["sex"]=="Female"]
```

```python
print(female_df)
```

```python
avg_tip_male=male_df["tip"].mean()
```

```python
avg_tip_female=female_df["tip"].mean()
```

```python
print(avg_tip_male)
```

```python
print(avg_tip_female)
```

```python
avg_total_bill_male = male_df["total_bill"].mean()
avg_total_bill_female = female_df["total_bill"].mean()
```

```python
ans = pd.DataFrame({
    "total_Bill":[avg_total_bill_male, avg_total_bill_female],
    "tip":[avg_tip_male, avg_tip_female]},
    index=["Male", "Female"]
)
```

```python
ans
```

#### hence total bill paid by male is greater than female

```python
for label, df in tips.groupby("sex"):
    print(label)
    print("mean bill:-", df["total_bill"].mean())
    print("mean tips:-", df["tip"].mean())
    print('-----'*30)
```

```python
tips.groupby("sex")  # tarnsformation
```

```python
tips.groupby("sex").agg({                            # aggregation
    "total_bill":["mean", "std", "min", "max"], 
    "tip":["mean", "std", "min", "max"],
    "size":["mean", "std", "min", "max"],
}).T
```

```python
tips.groupby(["sex", "smoker"]).agg({                            # aggregation
    "total_bill":["mean", "std", "min", "max"], 
    "tip":["mean", "std", "min", "max"],
    "size":["mean", "std", "min", "max"],
}).T
```
