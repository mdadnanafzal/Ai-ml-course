```python
import numpy as np
import matplotlib.pyplot as plt
```

Data Distribution
1. Uniform Distribution
2. Normal Distribution(Gaussian Distribution)

shape of data/distribution

```python
data = np.random.randint(1, 100, 1000)
# uniform distribution
plt.figure(figsize=(8, 3))
plt.hist(data, bins = 30)
plt.show()
```

```python
height = np.random.normal(1000)
# standard normal distribution mean = 0, std =1()
plt.figure(figsize=(8, 3))
plt.hist(height, bins=30, ec='k', label="stanadard normal distribution")
plt.legend()
plt.show()
```

```python
height = np.random.normal(160, 15, 1000)
# standard normal distribution mean = 0, std =1()
plt.figure(figsize=(8, 3))
plt.hist(height, bins=30, ec='k', label="stanadard normal distribution")
plt.legend()
plt.show()
```

```python
new_data = np.append(height, np.random.uniform(100, 130, 50))
```

```python
plt.hist(new_data, bins=30, label="left skewed normal distribution")
plt.show()
```
