### conditionals

```python
df[df["height"] > 170]
```

```python
df[df["gender"] == "male"]
```

```python
df[(df["height"] > 170) & (df["gender"] =="female") & (df["age"] > 20)]
```

```python
df.loc[(df["height"] > 170) & (df["gender"] =="female") & (df["age"] > 20),
       ["height","age"]
      ]
```
